```python
import pandas as pd
import json

# Leer el archivo JSON completo
with open('data/starwars_character_details.json') as archivo:
    datos_json = json.load(archivo)

# Extraer los valores dentro de la llave 'fields'
valores_fields = [elemento['fields'] for elemento in datos_json]

# Convertir los valores extraídos en un DataFrame de Pandas
df = pd.DataFrame(valores_fields)

# Mostrar el contenido del DataFrame
df
```

```python
df[['date_edited', 'time_edited_utc']] = df['edited'].str.replace('Z', '').str.split('T', expand=True)
df[['date_created', 'time_created_utc']] = df['created'].str.replace('Z', '').str.split('T', expand=True)
```

```python
df['date_edited'] = pd.to_datetime(df['date_edited']).dt.date
df['time_edited_utc'] = pd.to_datetime(df['time_edited_utc']).dt.time
df['date_created'] = pd.to_datetime(df['date_created']).dt.date
df['time_created_utc'] = pd.to_datetime(df['time_created_utc']).dt.time
```

```python
df.date_created.value_counts().sort_index()
```

```python
df.drop(columns=['edited','created'],inplace=True)
```

```python
df.to_csv('data/starwars_character_details.csv')
```
